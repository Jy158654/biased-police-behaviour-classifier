```python
sanfrancisco.describe(include="all")
```

From the describe function, there are still quite alot of data with no values (NA). However, it is important to note that not all NaN means no data. 

In the case of district, subject_age, search_vehicle, reason_for_stop", NaN does imply a an absence of data. But for contraband_found in particular, NaN actually means a search was not conducted, and in the case of outcome, NaN means that nothing happened to the driver that was stopped, not that the data is imcomplete. Therefore, special care must be taken to not accidentally remove these row.

```python
#remove rows with null values except contraband_found
sanfrancisco = sanfrancisco.dropna(axis=0, subset=["district", "subject_age", "search_vehicle", "reason_for_stop"])
sanfrancisco.describe(include="all")
```

```python
#replace NaN with more meaningful information
sanfrancisco["contraband_found"].fillna(value="search not conducted", inplace=True)
sanfrancisco["outcome"].fillna(value = "without valid reason", inplace = True)
```

```python
sanfrancisco.describe(include="all")
```

```python
sanfrancisco.isna().sum()
```

```python
sanfrancisco.reset_index(drop=True, inplace=True)
```

As for the colummn reason_for_stop, there are instances where a driver could be stopped for more that one reason. As such, we have chosen to use the first entry as the main reason for stop to minimise the unique entries to simplify the analysis. For cases where NA exists, the next reason will be used.

```python
sanfrancisco["reason_for_stop"].value_counts()
```

```python
sanfrancisco["reasons_for_stop"] = sanfrancisco["reason_for_stop"]

sanfrancisco["no_of_reasons"] = ""

for i in range(sanfrancisco.shape[0]):
    count = 1
    for j in sanfrancisco.loc[i, "reasons_for_stop"]:
        if j == "|":
            count += 1
    sanfrancisco.loc[i, "no_of_reasons"] = count
```

```python
sanfrancisco.describe(include="all")
```

```python
# choose the 1st reason not equals to NA for stops that are stopped for more than 1 reason
count = 0
sanity_check_df = []
for i in range(sanfrancisco.shape[0]):
    temp = str(sanfrancisco.loc[i, "reason_for_stop"]).split("|")
    num_of_categories = len(temp)
    if (num_of_categories > 1):
        if temp[0] != "NA":
            sanfrancisco.loc[i, "reason_for_stop"] = str(temp[0])
        else:
            sanfrancisco.loc[i, "reason_for_stop"] = str(temp[1])
        sanity_check_df.append(sanfrancisco.loc[i, ])
        count +=1
print("There are", count, "stops that are stopped for more than 1 reason.")
sanity_check_df = pd.DataFrame(sanity_check_df)
sanity_check_df.shape

print("\n\nEntries after cleaning:")

sanfrancisco["reason_for_stop"].value_counts()
```

# Feature Engineering

##### Categorical: district, subject_race, subject_sex, outcome, contraband_found, search_vehicle, reason_for_stop
##### Numerical： subject_age

### Creating Age Band

```python
#checking the age range
print("Min: ", min(sanfrancisco["subject_age"]) )
print("Max: ", max(sanfrancisco["subject_age"]))
```

```python
#adding column ageband
sanfrancisco["ageband"] = ""

sanfrancisco.loc[(sanfrancisco["subject_age"] < 15), "ageband"] = "children"
sanfrancisco.loc[(sanfrancisco["subject_age"] >= 15) & (sanfrancisco["subject_age"] < 25), "ageband"] = "youth"
sanfrancisco.loc[(sanfrancisco["subject_age"] >= 25) & (sanfrancisco["subject_age"] < 65), "ageband"] = "adult"
sanfrancisco.loc[(sanfrancisco["subject_age"] >= 65), "ageband"] = "senior"
```

### Grouping Outcome As "with valid reason"

```python
sanfrancisco["raw_outcome"] = sanfrancisco["outcome"]
sanfrancisco["outcome"] = sanfrancisco["outcome"].replace(["warning", "citation", "arrest"], "with valid reason")
```

```python
# convert date into standard format

sanfrancisco["date"] = sanfrancisco["date"].apply(lambda x: pd.to_datetime(x))
```

```python
for i in range(sanfrancisco.shape[0]):
    if "Moving Violation" in sanfrancisco.loc[i, "reasons_for_stop"]:
        sanfrancisco.loc[i, "m_v"] = True
    else:
        sanfrancisco.loc[i, "m_v"] = False
    if "Mechanical or Non-Moving Violation (V.C.)" in sanfrancisco.loc[i, "reasons_for_stop"]:
        sanfrancisco.loc[i, "mnm_v"] = True
    else:
        sanfrancisco.loc[i, "mnm_v"] = False
    if "MPC Violation" in sanfrancisco.loc[i, "reasons_for_stop"]:
        sanfrancisco.loc[i, "mpc_v"] = True
    else:
        sanfrancisco.loc[i, "mpc_v"] = False
    if "Traffic Collision" in sanfrancisco.loc[i, "reasons_for_stop"]:
        sanfrancisco.loc[i, "trafficol"] = True
    else:
        sanfrancisco.loc[i, "trafficol"] = False
    if "BOLO/APB/Warrant" in sanfrancisco.loc[i, "reasons_for_stop"]:
        sanfrancisco.loc[i, "b_a_w"] = True
    else:
        sanfrancisco.loc[i, "b_a_w"] = False
    if "DUI Check" in sanfrancisco.loc[i, "reasons_for_stop"]:
        sanfrancisco.loc[i, "dui_check"] = True
    else:
        sanfrancisco.loc[i, "dui_check"] = False
    if "Assistance to Motorist" in sanfrancisco.loc[i, "reasons_for_stop"]:
        sanfrancisco.loc[i, "assistance"] = True
    else:
        sanfrancisco.loc[i, "assistance"] = False
```
